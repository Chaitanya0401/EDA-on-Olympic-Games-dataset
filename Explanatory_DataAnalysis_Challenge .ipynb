## Summer Games vs. Winter Games - does Geographical Location matter?

__Identify Countries__ that are 
1. __equally successful__ in Summer and Winter Games
2. more successful in __Summer Games__
3. more successful in __Winter Games__

```python
SumVsWin = pd.crosstab(olympics.Country, olympics.Edition)
```

```python
SumVsWin
```

```python
MenVsWomen = pd.crosstab(olympics.Country, olympics.Gender)
```

```python
df3= pd.concat([SumVsWin, MenVsWomen], axis = 1)
```

```python
df3
```

```python
df3["Total"]=df3.Summer+df3.Winter
```

```python
df3.sort_values(by="Total", ascending=False, inplace=True)
```

```python
ranks=df3.rank(axis=0, method="average", ascending=False)
ranks
```

```python
diff=ranks.Summer.sub(ranks.Winter).sort_values().to_frame()
```

```python
diff
```

```python
plt.figure(figsize = (40, 5))
sns.heatmap(diff.T,cmap='RdBu', center = 0)
```

## Men vs. Women - does Culture & Religion matter?

__Identify Countries__ where

## Do Traditions matter?

Create the __Seaborn Heatmap__ that shows the __Ranks__ of Top 50 Countries __by Sports__. <br>

```python
rankBySports = pd.crosstab(olympics.Country, olympics.Sport,margins=True).sort_values(by="All", ascending=False)
```

```python
rankBySports=rankBySports.rank(method="average", ascending=False)
```

```python
rankBySports=rankBySports.head(20)
```

```python
rankBySports.drop(index=["All"],columns=["All"], inplace=True)
```

```python
plt.figure(figsize = (30, 10))
sns.heatmap(rankBySports,cmap='RdYlGn_r', vmin = 1, vmax = 6, linewidth = 1)
plt.show()
```
